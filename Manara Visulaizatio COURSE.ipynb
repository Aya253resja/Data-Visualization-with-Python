```python
import matplotlib.pyplot as plt 

months=['Jan','Feb','Mar','Apr','May']
sales_productA =[50, 60, 70, 65, 80]
sales_productB =[40, 55, 60, 70, 75]
sales_productC =[30, 45, 55, 60, 65]

plt.plot(months, sales_productA, label='Product A', color='blue', marker='o')
plt.plot(months, sales_productB, label='Product B', color='red', marker='s')
plt.plot(months, sales_productC, label='Product C', color='green', marker='^')

plt.legend()

plt.title('Monthly products sales')
plt.xlabel('Months')
plt.ylabel('unit label')

plt.show()

```

```python
import matplotlib.pyplot as plt 
Categories=['A','B','C','D']
Value=[50,80,70,40]

plt.bar(Categories , Value , color='Darkblue' , edgecolor='black')
plt.title('Categories values')
plt.xlabel('Categories')
plt.ylabel('Value')

plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#to get data
tips=sns.load_dataset('tips')

sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.title('total_bills vs tip')
plt.xlabel('total_bills')
plt.ylabel('tip')

plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set2')
plt.title(' Distribution total_bill/day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```
